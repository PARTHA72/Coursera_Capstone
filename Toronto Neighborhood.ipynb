```python
#All Imports goes in this cell
from bs4 import BeautifulSoup
import requests # library to handle requests
import json # library to handle JSON files
import pandas as pd
```

```python
# Create an empty Panda Dataframe with the columns
column_names = ['PostalCode',  'Borough', 'Neighborhood' ] 

# instantiate the dataframe
TorontoNeighborhoods = pd.DataFrame(columns=column_names)
#TorontoNeighborhoods
```

##Assumptions made based on the HTML structure returned
1. The Entire Neighrbourhoor data is contained in a tag "Table". This part is at first secured via a find
2. Within the Table above there are td and tr representing rows and columns of the table
3. Each td has the Postal code within a <b> tag
4. Each td has a number of <a> tag
5. The first <a> has the Borough information under the 'title' attribute
6. The next few <a> has the Neighbourhood information under the 'title' attribute

```python
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
```

```python

source = requests.get(url).text
soup = BeautifulSoup (source, 'lxml')

#match = soup.title.text
TorontoFSATable = soup.find('table')
#print(TorontoFSATable)
RowDatas = TorontoFSATable.find_all('tr')

for RowD in RowDatas[0:] :
    ColumnDatas = RowD.find_all('td')
    for ColD in ColumnDatas[0:] :
    #Initialize the variables
        PostalCode=""
        Borough = ""
        Neighborhood=""
        #print (ColD.prettify())
        PostalCode = ColD.find('b').text
        BorNeigh = ColD.find_all('a')
        try :
            Borough = BorNeigh[0]['title']

            for Neigh in BorNeigh[1:] :
                Neighborhood = Neighborhood + Neigh['title'] +", "

        except Exception as e :
            pass
        
        if Borough != "" :
            if Neighborhood == "" :
                Neighborhood = Borough
            else:
                 Neighborhood=Neighborhood[0:-2]   
            TorontoNeighborhoods = TorontoNeighborhoods.append({ 'PostalCode': PostalCode,
                                            'Borough': Borough,
                                            'Neighborhood': Neighborhood}, ignore_index=True)
        #print (PostalCode)




#print(ColumnData.prettify())
```
